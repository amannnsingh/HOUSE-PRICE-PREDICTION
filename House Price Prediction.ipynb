# REMOVAL OF DUPLICATE VALUE

```python
counter = 0
rs,cs = data.shape

data.drop_duplicates(inplace=True)

if data.shape==(rs,cs):
    print('\n\033[1mInference:\033[0m The dataset doesn\'t have any duplicates')
else:
    print(f'\n\033[1mInference:\033[0m Number of duplicates dropped/fixed ---> {rs-data.shape[0]}')
```

```python
from sklearn.preprocessing import LabelEncoder
categ = ["Avg. Area Income","Avg. Area House Age","Avg. Area Number of Rooms","Avg. Area Number of Bedrooms","Area Population","Price"]
```

```python
le = LabelEncoder()
data[categ] = data[categ].apply(le.fit_transform)
```

# Training a Linear Regression Model

```python
X = data[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
               'Avg. Area Number of Bedrooms', 'Area Population']]

y = data['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

# LinearRegression Model Evaluation

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

# Predictions from our Linear Regression Model

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

In the above scatter plot, we see data is in line shape, which means our model has done good predictions.

```python
sns.distplot((y_test-predictions),bins=50);
```

In the above histogram plot, we see data is in bell shape (Normally Distributed), which means our model has done good predictions.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

# Evaluation :

```python
from sklearn.metrics import r2_score,mean_absolute_error
score = r2_score(y_test,y_predict)
error = mean_absolute_error(y_test,y_predict)
```
